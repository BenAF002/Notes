# 1.1 Discussion: The Irrationality of $\sqrt{2}$

***Theorem 1.1.1***\
There is no rational number whose square is 2.
$$(p/q)^2 \ne 2, \ \forall p,q \in \mathbb{Z}$$

A ***Rational Number*** is any number that can be expressed in the form $p / q, \ p,q\in\mathbb{Z}$\
So, the theorem states that no rational number, squared, is equal to $2$.

*Proof:*\
This proof assumes that the theorem is false, then demonstrates that it cannot be false. It is a **Proof by Contradiction**

Assume: $\exists p,q \in \mathbb{Z} \ : \ (p/q)^2 = 2$\
This implies that: $p^2 = 2q^2$\
Thus $p^2$ must be an even integer, since $p$ and $q$ are both integers. Thus, we may substitute $p = 2r$ where $r$ is also an integer, and then express: $$2r^2 = q^2$$
This implies that $q$ *is also even*. Thus, $p$ and $q$ must both be divisible by $2$ and we may reduce their quotient to a lower common factor.\
 If this is true, then we can infinitely reduce them recursively. Thus, $(p/q)^2 = 2$ *cannot hold* for **any** integers $p$ and $q$.\
Thus proves the theorem.

**Field**:\
A *field* is any set on which the operations of addition and multiplication are:
- Commutative: $a + b = b + a$ (addition), $ab = ba$ (multiplication)
- Associative: $(a + b) + c = a + (b + c)$ (addition), $(ab)c = a(bc)$
- Distributive: $a(b+c) = ab+ ac$
- **AND** possesses additive identity and inverses **that are also in the set**
- **AND** possesses multiplicative identity and inverses **that are also in the set**

**The Rational Numbers:**
$$\mathbb{Q} = \{\text{all fractions } p / q \ : \ p,q\in\mathbb{Z}, \ q\ne 0\}$$

The integers $\mathbb{Z}$ **are not a field** because not all integers have a multiplicative inverse that *is also an integer*. E.g., the multiplicative inverse of $2$ is $1/2$ such that $2(1/2) = 1$. The number $1/2\notin\mathbb{Z}$, therefore the multiplicative inverse of $2$ is not in the set of integers $\mathbb{Z}$ and $\mathbb{Z}$ is *not a field*

The set of rational numbers $\mathbb{Q}$ *is a field*. It obeys all of the associative, distributive and commutative properties and has closure under the additive and multiplicative inverses of all of its consituents.

There are *no gaps* in the line of rational numbers. Take any two rational numbers $r$ and $s$. The midpoint between them *is also a rational number*: $$\frac{r + s}{2}$$

However, we know that $\sqrt{2}$ (along with heaps of other irrational numbers) *do not exist* on the line of rational numbers. We can infinitely approximate them in a sense, by repeating this midpoint calculation for instance, but we can never actually reach them...

**The Real Numbers:**

The real numbers $\mathbb R$ are obtained by "*filling in the gaps*" in the rational numbers with irrational numbers.

# 1.2 Preliminaries

### Sets
A set is a collection of elements.\
The text uses $A\subseteq B$ to denote that set $A$ is a subset of set $B$

Sets may contain sets. An infinite recursion of sets may be: $$A_1\supseteq A_2 \supseteq A_3 \supseteq \cdots$$
And a union of infinitely many sets is expressable as:
$$\bigcup_{n=1}^\infty A_n$$
Which for the infinite recursion above is $\bigcup_{n=1}^\infty A_n = A_1$\

**Complements**\
The complement of set $A$ is denoted $A^c$ and is the set of all elements in a superset containing both $A$ and $A^c$ that are not in $A$.\
 In a field (or space) like $\mathbb R$, the complement of a set $A\in \mathbb R$ is the set $A^c\in\mathbb R$ such that: $$A^c = \{x\in\mathbb R: x \notin A\}$$

**De Morgan's Laws:**
$$(A\cap B)^c = A^c \cup B^c$$
$$(A\cup B)^c = A^c \cap B^c$$

### Functions

**Definition 1.2.3**\
Given two sets $A$ and $B$, a *function* is a rule or mapping that takes each element $x\in A$ and associates it with *a single* element $f(x) \in B$
$$f: A\rightarrow B$$
For such a function $f$:
$$
\begin{align*}
& \text{Domain: } A \\
& \text{Range: } \{y\in B : f(x) = y\}
\end{align*}
$$

**Example 1.2.5 (Triangle Inequality)**\
$|x|$ denotes the *absolute value function* defined as: $$|x| = \begin{cases} x, \ x \ge 0 \\ -x, \ x < 0 \end{cases}$$
Triangle inequality: $$|a + b| \le |a| + |b|, \ \forall a,b$$

### Logic and Proofs

The steps in a proof must either follow logically from eachother or be justified by other agreed facts.

Theorem 1.1.1 was proven with a *proof by contradiction* which is an example of an ***Indirect Proof***.\
A ***Direct Proof*** follows from a valid statement, often from the theorem's hypothesis, and demonstrates the theorem's hypothesis through logical deductions

**Theorem 1.2.6**\
Two real numbers $a$ and $b$ are equal if and only if: $$|a-b|<\epsilon, \ \forall \epsilon>0 \in \reals$$

*Proof:*\
Because this theorem includes "if and only if", we need a ***bidirectional proof***, we need to prove that it is true in both directions:
- $(\rightarrow)$ if $a=b$, then for all real numbers $\epsilon > 0$, $|a-b|<\epsilon$
- $(\leftarrow)$ if $|a-b|<\epsilon$ for all real numbers $\epsilon >0$, then $a=b$

For the first direction, $a=b \implies a-b=0 \implies |a-b|=0 \implies |a-b|<\epsilon, \ \forall \epsilon >0 \in \reals$\
This is a direct proof.

The second direction may be proven by contradiction, so indirectly, by assuming that $a\ne b$ *and* the first clause of the statement is true.\
Thus we assume that **all** $\epsilon > 0$ in $\reals$ are such that $|a - b| < \epsilon$. Since, $a$ and $b$ are both in $\reals$, $|a - b|\in \reals$ and $a \ne b \implies |a - b| > 0$:
$$\therefore \exists \epsilon_0\in\reals \ : \ \epsilon_0 = |a - b| > 0$$
However, $$|a - b| < \epsilon_0 \ \text{ and } \ |a - b| = \epsilon_0$$ **cannot** both be true, they are *contradictions*.\
Therefore, $a\ne b$ is impossible and it must be the case that $a = b$

### Induction

Used in this book, induction arguments, or inductive proofs, prove a statement over the *natural numbers* $\mathbb N$ or sets by showing that the statement is true for a minimal base case, then that it is true for any arbitrary case $n$, after which it must be true for any next $n+1$ case. This proves that the statement must be true over all of $\mathbb N$.

Let $S\subseteq \mathbb N$ with the following properties:
- $1 \in S$
- Whenever $S$ contains a natural number $n$, it also contains $n+1$
If these two properties are satisfied, then it must be that $S = \mathbb N$

**Example 1.2.7**\
Let us define $x_n$ for all $n \in \mathbb N$ as:
$$
\begin{align*}
& x_1 = 1 \\
& x_{n+1} = (1/2)x_n + 1, \ \forall n \in \mathbb N
\end{align*}
$$

Let us prove by induction that the sequence $x_1, ..., x_n, ... $ is *increasing*.\
We have the base case by definition: $(x_1 = 1) \le (x_2 = 3/2)$\
Next, we need the arbitrary case: $x_n \le x_{n+1}, \ \forall n \in \mathbb N$:

Let's denote $S \coloneqq \{x_n \ : \ x_1 = 1, \ x_{n+1}=(1/2)x_n + 1, \ \forall n \in \mathbb N\}$\
Then, for *any* $x_n$ and $x_{n+1}$ in the set, the relationship holds that:
$$x_n \le x_{n+1} \implies (1/2)x_n + 1 \le (1/2)x_{n+1} + 1 \\ \ \\ \therefore x_n \le x_{n+1} \ \forall n \in \mathbb N$$

The final statement $\therefore$ follows from **induction**. The left side of the inequality being implied is the definition of $x_{n+1}$, thus we may apply this inequality to $x_1 \le x_2$, which we know to be true, to yield $x_2\le x_3$. We may do this recursively to show that *any* $x_n\le x_{n+1}$.\
Thus proves the statement.

# 1.3 The Axiom of Completeness

We stated above that $\reals$ is an extension of the rational numbers $\mathbb Q$ that *fills in the gaps* in $\mathbb Q$.

### An Initial Definition of $\reals$

$\reals$ is a *set* containing $\mathbb Q$.\
Every element in $\reals$ has an additive inverse and every non-zero element in $\reals$ has a multiplicative inverse.\
$\reals$ is a *field* - so addition and multiplication are *associative*, *commutative* and *distributive*

**Axiom of Completeness**\
Every *nonempty* set of real numbers that is bounded *above* has *at least and upper bound*

### Least Upper Bounds and Greatest Lower Bounds

**Definition 1.3.1**\
A set $A\subseteq \reals$ is ***Bounded Above*** if:
$$ \exists b\in \reals \ : \ a\le b, \ \forall a \in A$$
Such a $b$ is called an *upper bound* for $A$

A set $A$ is ***Bounded Below*** if:
$$ \exists l\in \reals \ : \ l\le a, \ \forall a \in A$$

**Definition 1.3.2**\
A real number $s$ is the ***Least Upper Bound*** (aka. *supremum*) for a set $A\subseteq \reals$ if it meets these two criteria:
1. $s$ is an *upper bound* for $A$
2. Any other upper bound $b$ for $A$ is such that: $s\le b$

$s$ is often denoted $s\coloneqq \text{sup} A$

The ***Greatest Lower Bound*** (aka. *infimum*) for $A$ is similarly defined, and is denoted $\text{inf} A$

**IMPORTANT NOTE:** $\text{sup} A$ and $\text{inf} A$, may *not* be elements of $A$

**Definition 1.3.4**\
A real number $a_0$ is a ***Maximum*** of a set $A$ if $a_0 \in A$ and $a_0 \ge a, \ \forall a\in A, a\ne a_0$\
A real number $a_1$ is a ***Minimum*** of a set $A$ if $a_1 \in A$ and $a_1 \le a, \ \forall a\in A, a\ne a_1$

**Example 1.3.5**\
$A=\{x\in(0, 2)\}$, i.e. $\{x\in \reals \ : \ 0 < x < 2\}$, **does not** have a maximum. It **does** have a supremum of $2$\
$A=\{x\in[0, 2]\}$, i.e. $\{x\in \reals \ : \ 0 \le x \le 2\}$, **does** have a maximum of $2$. It **also** has a supremum of $2$\
If a set has a maximum, then the maximum is also its supremum. The same goes for minimums and infimums.

Although not every bounded set has a maximum, the Axiom of Completeness does assert that every non-empty bounded set **does have a least upper bound**, and likewise for greatest lower bound.
